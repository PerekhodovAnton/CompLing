Реализуйте алгоритм Symspell. Он похож на алгоритм Норвига, но проще и быстрее. Он основан только на одной операции - удалении символа. Описание алгоритма по шагам:

1) Составляется словарь правильных слов  
2) На основе словаря правильных слов составляется словарь удалений - для каждого правильного слова создаются все варианты удалений и создается словарь, где ключ - слово с удалением, а значение - правильное слово  (!)
3) Для выбора исправления для слова с опечаткой генерируются все варианты удаления, из них выбираются те, что есть в словаре удалений, построенного на шаге 2. Слово с опечаткой заменяется на правильное слово, соответствующее варианту удаления  
4) Если в словаре удалений есть несколько вариантов, то выбирается удаление, которому соответствует наиболее вероятное правильное слово

```python
# Находим только уникальные слова
true_only = []
for sent in true:
  for word in sent.split():
    true_only.append(word)

true_only_singles = list(set(true_only))
true_only_singles
```

```python
# Создаем словарь, где ключь - уникальное слово (из того, что выше), а значение - каждый вариант удаления одной буквы из слова
dictionary = {}
for word in true_only_singles:
  count = 0
  words_deleted = []

  while count < len(word):
    """
    здесь я потратил много времени, потому что изначально использовал replace, но позже понял, что,
    несмотря на то, что метод позволяет выбрать сколько повторов букв удалить, он не понимает их порядок.
    Из-за этого из слова люблю два разва удалялась именно первая "ю"
    Пофиксил это таким образом:
    """
    corrections = word[:count] + word[count + 1:]
    words_deleted.append(corrections)
    count += 1

  dictionary.update({word : words_deleted})
dictionary
```

```python
# первая функция проверяет, что мб слово уже есть в словаре и тогда сразу говорим True. Или отдаем слово для след. фунции
def in_dict_try(word:str):
  for w in dictionary.keys():
    if w == word:
      return True

  return word

# Вот тут самый сок, очень долго его придумывал, хотя сейчас кажется, что все очень просто и понятно...
def compare(word):
  if len(word) > 1:
    # проверяем по функции выше
    if in_dict_try(word) != True:
      counter1 =  Counter(true_only)
      compare = {}
      #  сравниваем слово с каждым значением ключа
      for w in dictionary:
        for k in dictionary[w]:
          if len(word) > 1:
            if word == k:
              #  если все гуд, то отдаем в словарь частоту вхождения из главного словаря и слово-ключ
              compare[w] = counter1[w], word

      try:
        # тут выбираем самое вероятное слово. Вероятное = самое частотное
        return max(compare)
      except:
        return False

    if in_dict_try(word) == True:
      return word

  else:
    return word
```

```python
# вот как выглядит работа алгоритма. На вход можно предложение или слово

def spell_check(sentence:str):
  sent = [i for i in sentence.split()]
  if len(sent) > 1:
    corrected_sent = []
    for word in sent:
      if compare(word) != False:
        corrected_sent.append(compare(word))
      else:
        corrected_sent.append(word)

    return ' '.join(corrected_sent)
  else:
    return(compare(sent[0]))
```

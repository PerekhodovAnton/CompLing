## 3. Лемматизация Mystem vs Pymorphy (2 балла)

Лемматизируйте текст с помощью mystem и pymorphy. Найдите различия в лемматизации. Что по вашему работает лучше на приведенном тексте?

Важно: для пайморфи используйте токенизацию из mystem, чтобы исключить влияние токенизации на результат. Анализируйте только значимые различия, а не технические особенности (не сравнивайте скорость работы и удобность интерфейса).

```python
pip install pymorphy2
```

```python
from pymorphy2 import MorphAnalyzer
morph = MorphAnalyzer()

morph_lemmas = []
for i in mystm:
  m = morph.parse(i)
  morph_lemmas.append(m[0].normal_form)

mstem_lemmas = mystem.lemmatize(text)
mstem_lemmas = [i for i in mstem_lemmas if i != ' ']

morph_diff = [ph for ph in morph_lemmas if ph not in mstem_lemmas]
mstem_diff = [m for m in mstem_lemmas if m not in morph_lemmas]

split_text = text.lower().split()

print(f'mstem_diff : {mstem_diff} \n morph_diff : {morph_diff} \n morph_lemmas : {morph_lemmas} \n mstem_lemmas : {mstem_lemmas} \n text : {split_text}')

# мне кажется, что mystem работает лучше, т.к. корректнее определяет начальную форму, но оба варианта плохо работаю с числами
```

## 4. Лемматизация в SpaCy (2 балла)

С помощью Spacy (модель для русского языка) лемматизируйте тот же текст. Проверьте есть ли различия с Mystem и Pymoprhy.

```python
import spacy

nlp = spacy.load("ru_core_news_sm")
doc = nlp(text)
```

```python
spacy_lemmas = []
for sent in doc.sents:
    for token in sent:
        spacy_lemmas.append(token.lemma_)


spacy_diff_mstem = [s for s in spacy_lemmas if s not in mstem_lemmas]
spacy_diff_morph = [s for s in spacy_lemmas if s not in morph_lemmas]

print(f' spacy_diff_mstem : {spacy_diff_mstem} \n spacy_diff_morph : {spacy_diff_morph} \n spacy_lemmas : {spacy_lemmas} \n text : {split_text}')

# есть различия в выдаваемых леммах, у mystem они лучше всех
```
